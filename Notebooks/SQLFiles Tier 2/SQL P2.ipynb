```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
con = engine.connect()
rs = con.execute('SELECT firstname, name, monthlymaintenance, starttime, joindate, SUM(CASE WHEN firstname <> "GUEST" THEN membercost ELSE guestcost END)-monthlymaintenance AS total_cost FROM Bookings AS b LEFT JOIN Members AS m ON b.memid = m.memid LEFT JOIN Facilities AS f ON b.facid = f.facid GROUP BY name HAVING total_cost > 1000')
df= pd.DataFrame(rs.fetchall())
df.columns = rs.keys()
con.close()
df
```
